```python
import pandas as pd
import glob
from openpyxl import load_workbook

# Directory where your Excel files are stored
directory_path = "data_kafka20lac/*.xlsx"

# Find all Excel files in the directory
excel_files = glob.glob(directory_path)

# Initialize an empty DataFrame to store all data
final_df = pd.DataFrame()

# Function to read data from openpyxl and append to DataFrame
def read_excel_openpyxl(file):
    wb = load_workbook(filename=file, data_only=True)  # Load the Excel workbook
    all_data = []

    for sheet_name in wb.sheetnames:  # Loop through all sheets
        ws = wb[sheet_name]
        # Extract sheet data row by row
        sheet_data = [[cell.value for cell in row] for row in ws.iter_rows(values_only=True)]
        
        # Create a DataFrame for the sheet and append to final_df
        df = pd.DataFrame(sheet_data[1:], columns=sheet_data[0])  # First row is header
        all_data.append(df)
    
    return pd.concat(all_data, ignore_index=True)

# Loop through each Excel file
for file in excel_files:
    # Read data from each Excel file and append it to final_df
    file_data = read_excel_openpyxl(file)
    final_df = pd.concat([final_df, file_data], ignore_index=True)

# Now, final_df contains data from all Excel sheets in all files
```

```python

import glob

# Directory where your Excel files are stored
directory_path = "data_kafka20lac/*.xlsx"

# Find all Excel files in the directory
excel_files = glob.glob(directory_path)

# Initialize an empty list to store data from all sheets
all_data = []

# Loop through each Excel file
for file in excel_files:
    # Read all sheet names from the Excel file
    xls = pd.ExcelFile(file)
    
    # Loop through each sheet and append the data
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(file, sheet_name=sheet_name)
        all_data.append(df)

# Concatenate all data into a single DataFrame
final_df = pd.concat(all_data, ignore_index=True)

# Now, final_df contains data from all Excel sheets in all files
```

```python
def remove_code(raw_msg):
        
        if raw_msg != '':
            index = raw_msg.find('>')
            msg_str = raw_msg[index+1:]
            return msg_str
        else:
            return ''  
```

```python
import pandas as pd
import ast

# Function to parse the 'value' column and create a shortened message
def extract_short_message(value):
    try:
        # Convert string to dictionary
        value_dict = ast.literal_eval(value)  # Safely evaluate string as dictionary
        payload = value_dict.get('payload', '')
        short_message = remove_code(payload)
        return short_message
    except Exception as e:
        # Handle errors (e.g., malformed data)
        return None
    
```

```python
# Apply the extraction function to the 'value' column to create the 'message' column
df['message'] = df['value'].apply(extract_short_message)
```

```python
date_pattern = r"(\w+)\s+(\d+)\s+(\d+)\s+(\d+:\d+:\d+)"
host_pattern = r"(\S+)\s+%%"
module_pattern = r"%%(\d+)(\w+)/(\d+)/(\w+)"
description_pattern = r'(?P<description>.*)$'
```

```python
months, days, years, times, hosts = [], [], [], [], []
version_numbers, module_names, severities, log_types = [], [], [], []
descriptions = []
```

```python
for i in range(len(df)):
    
    log_entry = df['message'].iloc[i]

    
    month, day, year, time = None, None, None, None
    host = None
    version_number, module_name, severity, log_type = None, None, None, None
    description = None


    
    date_match = re.search(date_pattern, log_entry)
    host_match = re.search(host_pattern, log_entry)
    module_match = re.search(module_pattern, log_entry)
    description_match = re.search(description_pattern, log_entry)

    
    if date_match:
        month, day, year, time = date_match.groups()
    if host_match:
        host = host_match.group(1)
    if module_match:
        version_number, module_name, severity, log_type = module_match.groups()
    if description_match:
        description = description_match.group()

    
    months.append(month)
    days.append(day)
    years.append(year)
    times.append(time)
    hosts.append(host)
    version_numbers.append(version_number)
    module_names.append(module_name)
    severities.append(severity)
    log_types.append(log_type)
    descriptions.append(description)
```

```python
# Filter data for the specific host and date
selected_host = 'GPN-0004_CHB-CR_M14-01'
day = '30'
month = 'Sep'
filtered_data = df[(df['host'] == selected_host) & (df['day'] == day) & (df['month'] == month)]

# filtered_data['timstap_index_1'] = filtered_data['timestamp']
# Set the timestamp as the index
filtered_data.set_index('timestamp', inplace=True)

# Resample to get counts of logs per hour
hourly_counts = filtered_data.resample('H').count()

# Count failed logs
failed_counts = filtered_data[filtered_data['message'].str.contains('down', case=False)].resample('H').count()

# Step 5: Align indices and fill missing values in failed_counts with 0
failed_counts = failed_counts.reindex(hourly_counts.index, fill_value=0)

# Step 6: Calculate success logs as total minus failed logs
success_counts = hourly_counts - failed_counts
```

```python
# Step 7: Plot the stacked bar chart (green for success, red for failure)
plt.figure(figsize=(10, 6))

# Stacked bar chart for failed and successful logs
plt.bar(hourly_counts.index, success_counts['message'], bottom=failed_counts['message'], color='green', label='Success', width=0.03)
plt.bar(hourly_counts.index, failed_counts['message'], color='red', label='Failure', width=0.03)

# Customize the graph
plt.title('Hourly Log Activity (Success vs Failure)')
plt.xlabel('Hour')
plt.ylabel('Number of Logs')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
